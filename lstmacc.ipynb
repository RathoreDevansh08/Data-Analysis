```
# IMPORTS
import numpy as np
from keras.models import Sequential
from keras.layers import LSTM
from keras.layers import Dense
```

```
x_width = 3
```

```
#Data Load
xtrain = [[i+j for j in range(x_width)]for i in range(100)]#created list
ytrain = [[i+j for j in range(0+2,x_width+2)]for i in range(100)] #adding 2

xtest= [[i+j for j in range(3)]for i in range(100,200)]
ytest = [[i+j for j in range(0+2,x_width+2)]for i in range(100,200)]

# print(type(xtrain))         #understand
# print(np.shape(xtrain))
# print(xtrain)

# print(type(np.array(xtrain, dtype=np.float32)))
# print(np.shape(np.array(xtrain, dtype=np.float32)))
# # print(np.array(xtrain, dtype=np.float32))

xtrain = np.array(xtrain, dtype=np.float32).reshape(100,1,3)/100#converting list to array, we cant reshape list, and then reshaping it. 
ytrain = np.array(ytrain, dtype=np.float32).reshape(100,1,3)/100
xtest  = np.array(xtest, dtype=np.float32).reshape(100,1,3)/100
ytest  = np.array(ytest, dtype=np.float32).reshape(100,1,3)/100

# print(type(xtrain))
# print(np.shape(xtrain))
# print(xtrain)
```

```
layer_width = x_width
model = Sequential()
LSTM_layer1 = model.add(LSTM(layer_width, input_shape = (1,layer_width), return_sequences = True, activation = 'sigmoid'))
LSTM_layer2 = model.add(LSTM(layer_width, input_shape = (1,layer_width), return_sequences = True, activation = 'sigmoid'))
LSTM_layer3 = model.add(LSTM(layer_width, input_shape = (1,layer_width), return_sequences = True, activation = 'sigmoid'))
LSTM_layer4 = model.add(LSTM(layer_width, input_shape = (1,layer_width), return_sequences = True, activation = 'sigmoid'))
LSTM_layer5 = model.add(LSTM(layer_width, input_shape = (1,layer_width), return_sequences = True, activation = 'sigmoid'))
LSTM_layer6 = model.add(LSTM(layer_width, input_shape = (1,layer_width), return_sequences = True, activation = 'sigmoid'))
#dense_layer = model.add(Dense(3))
model.compile(loss = 'mse', optimizer = 'adam', metrics = ['accuracy'])
model.fit(xtrain, ytrain, nb_epoch = 10000, batch_size = 50, validation_data = (xtest, ytest) )
```

```
predict = model.predict(xtrain)*100
print(predict)
```
