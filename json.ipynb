```
import spacy
en_nlp = spacy.load('en')
```

```
Label = []

for k in range(0,len(datastore)):
  for x in range(0,len(datastore[k]['profStudy'])):
    tempdoc = en_nlp(datastore[k]['profStudy'][x])
    for ent in tempdoc.ents:
      Label.append(ent.label_)  
 


Label = np.unique(Label)
n_labels = len(Label)
print(n_labels)
print(Label)


entlist = []
for i in range(0,n_labels):
  entlist.append(i)
  entlist[i] = []
temp = []

for k in range(0,len(datastore)):
  for x in range(0,len(datastore[k]['profStudy'])):
    tempdoc = en_nlp(datastore[k]['profStudy'][x])
    for ent in tempdoc.ents:
      for i in range(0,n_labels):
        if ent.label_ == Label[i]:
          entlist[i].append(ent.text)
          break;
   
  
for i in range(0,n_labels):
  entlist[i] = np.unique(entlist[i])
  
  
for i in range(0,n_labels):
  print('\n' + Label[i] + ':')
  print(entlist[i])
```

```
ulist = ["school", "university", "institute"]

print("------------------>profStudy<---------------------------\n")
for k in range(0,len(datastore)):
  print("\n" + str(k) + ") " +str(datastore[k]['profStudy']))
  for x in range(0,len(datastore[k]['profStudy'])):
    tempdoc = en_nlp(datastore[k]['profStudy'][x])
    print("\n  ->" + str(tempdoc))
    flag = 0
    for ent in tempdoc.ents:
      if ent.label_ == "DATE":
        print("     -DATE:" + ent.text)
        flag = 1
      
    for token in tempdoc:
      for y in range(0,7):
        if str(token) == newstudylist[y] or str(token) == newstudylist[y+8] or str(token) == newstudylist[y+16]:
          print("     -Degree:" + newstudylist[y])
          flag = 1

    for i,phrase in enumerate(str(tempdoc).split(',')):
      for token in en_nlp(phrase):
        if token.lower_ in ulist:
          print("     -University:" + phrase)
          
        
        
          flag1 = 0
          for token in en_nlp(str(tempdoc).split(',')[i-1]):
            if token.text in newstudylist:
              flag1 = 1    
          if flag1 == 0:
            print("     -Study:" + str(tempdoc).split(',')[i-1])
          
          print("     -Location:", end =" ")
          for phrase in str(tempdoc).split(',')[i+1:]:
            tempdoc1 = en_nlp(phrase)
            flag2=0
            for ent in tempdoc1.ents:
              if ent.label_ == "DATE":
                flag2=1
            if flag2 == 0:
              print(phrase + ",")
```
